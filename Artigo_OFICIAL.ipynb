```python
#regressão com dois eixos #NESSE GRÁFICO ESTOU USANDO DUAS PLANILHAS DIFERENTES USEI UM DATA FRAME PARA CADA UMA
#https://gist.github.com/endolith/2719900   core

#lendo os arquivos
#nesse caso como não pode ter células vazias e foram tirados alguns dados, é necessario separar em planilhas
arquivo=pd.ExcelFile('_dados_tratados.xlsx')
data=pd.read_excel(arquivo,'Correlation')


#SIF 757
points = plt.scatter(y='XCO2', x='SIF757',data=data,
                     c=data["SIF757"],marker="8", s=80, cmap="summer")  #cmap é a cor da barra  
cbar = plt.colorbar(points, shrink=1)
cbar.set_label('SIF 757',fontsize=12, family='Segoe UI')  #COLOCAR O NOME DA BARRA
sns.regplot(y='XCO2', x='SIF757', data=data, scatter=False, color=".1")

#SIF 771
points = plt.scatter(y='XCO2', marker="8", x='SIF771',data=data,
                     c=data["SIF771"], s=80 ,label=False, cmap="BuGn")
cbar =plt.colorbar(points, shrink=1)
cbar.set_label('SIF 771',fontsize=12, family='Segoe UI')

sns.regplot(y='XCO2', x='SIF771', data=data,label=True, scatter=False, color=".1")


#formatando o gráfico e adicionandos os valores de p e R2 
#plt.text(1.12, 395, 'SIF771', fontsize=12) ADICIONANDO TEXTO NO GRÁFICO
r_squaredsSIF757 = 29.1
r_squaredsSIF771 = 57.6
pvalue = 0.001
plt.text(0.19, 398, 'R$\mathregular{^{2}}$ SIF 771= %0.1f' % r_squaredsSIF771,family='Segoe UI', fontsize=12)
plt.text(0.19, 397.7, 'R$\mathregular{^{2}}$ SIF 757= %0.1f' % r_squaredsSIF757,family='Segoe UI',fontsize=12)
plt.text(0.19, 397.4,'pvalue = <%0.3f' % pvalue,family='Segoe UI',fontsize=12)
plt.xlabel('SIF (Wm$\mathregular{^{-2}}$ sr$\mathregular{^{-1}}$ µm$\mathregular{^{-1}}$)',fontsize=12, family='Segoe UI')
plt.ylabel('XCO$\mathregular{_{2}}$(ppm)',fontsize=12, family='Segoe UI')
plt.rcParams['figure.figsize'] = (7,8)
plt.savefig('REGRESSAO_sif.png', transparent = False,bbox_inches='tight', dpi=1000)
```
